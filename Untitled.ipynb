```python
# Chapter 15: Sorting Algorithms

"""
Introduction
In this chapter, we will explore different sorting algorithms. Sorting is a fundamental operation in computer science, with applications ranging from database management to data analysis.

Concept: Sorting is important for efficient data retrieval, understanding data trends, and preprocessing data for further analysis. Sorting algorithms can be evaluated on properties like time complexity, space complexity, and stability.

Stability: A sorting algorithm is said to be stable if it preserves the relative order of records with equal keys. Stability is important in many real-world applications.
"""

import matplotlib.pyplot as plt
import numpy as np
import time

# Helper function to visualize sorting
def visualize_sorting(arr, title="Sorting Visualization"):
    plt.figure(figsize=(10, 5))
    plt.bar(range(len(arr)), arr, color="blue")
    plt.title(title)
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.show()

# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Testing Bubble Sort
array = [64, 25, 12, 22, 11]
visualize_sorting(array, title="Original Array for Bubble Sort")
sorted_array = bubble_sort(array.copy())
visualize_sorting(sorted_array, title="Bubble Sorted Array")

# Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Testing Selection Sort
array = [64, 25, 12, 22, 11]
visualize_sorting(array, title="Original Array for Selection Sort")
sorted_array = selection_sort(array.copy())
visualize_sorting(sorted_array, title="Selection Sorted Array")

# Insertion Sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Testing Insertion Sort
array = [64, 25, 12, 22, 11]
visualize_sorting(array, title="Original Array for Insertion Sort")
sorted_array = insertion_sort(array.copy())
visualize_sorting(sorted_array, title="Insertion Sorted Array")

# Merge Sort
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

# Testing Merge Sort
array = [64, 25, 12, 22, 11]
visualize_sorting(array, title="Original Array for Merge Sort")
sorted_array = merge_sort(array.copy())
visualize_sorting(sorted_array, title="Merge Sorted Array")

# Quick Sort
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Testing Quick Sort
array = [64, 25, 12, 22, 11]
visualize_sorting(array, title="Original Array for Quick Sort")
sorted_array = quick_sort(array.copy())
visualize_sorting(sorted_array, title="Quick Sorted Array")

# Heap Sort
import heapq

def heap_sort(arr):
    heapq.heapify(arr)
    sorted_arr = []
    while arr:
        sorted_arr.append(heapq.heappop(arr))
    return sorted_arr

# Testing Heap Sort
array = [64, 25, 12, 22, 11]
visualize_sorting(array, title="Original Array for Heap Sort")
sorted_array = heap_sort(array.copy())
visualize_sorting(sorted_array, title="Heap Sorted Array")

# Quiz
quiz = {
    1: "Which sorting algorithm has the best average time complexity for large datasets?",
    2: "What does it mean if a sorting algorithm is stable?",
    3: "Which sorting algorithm is the most efficient for small datasets?",
    4: "What is the difference between Merge Sort and Quick Sort?"
}

for q_no, question in quiz.items():
    print(f"Quiz {q_no}: {question}")

# Exercise
print("\nExercise: Implement a function that uses Merge Sort to sort a list of student scores and returns the sorted list.")

def exercise_merge_sort(scores):
    return merge_sort(scores)

# Example Usage
scores = [85, 67, 90, 78, 88, 76]
print("Sorted Scores:", exercise_merge_sort(scores))

# Solutions for Quiz and Exercise
"""
Quiz Solutions
1. Merge Sort and Quick Sort have the best average time complexity of O(n log n).
2. A sorting algorithm is stable if it preserves the relative order of elements with equal keys.
3. Insertion Sort is most efficient for small datasets.
4. Merge Sort divides the array into halves and merges them after sorting, whereas Quick Sort uses a pivot to partition and sort the array.

Exercise Solution
The solution for the exercise is already provided above using the merge_sort function.
"""
```
